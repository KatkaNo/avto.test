Сначала показалось, что задача на использование НОК или НОД (скорее НОК), но дело в том, что общее кратное чисел 3 и 5 не обязательно должно быть наименьшим, да и общим тоже. Например, 8 шариков - не делится ни на 3, ни на 5, но при этом такую порцию получить можно.

Тогда можно думать в сторону разложения числа k на слагаемые с участием 3 и 5, с какими-то коэф-ми: та же 8 = 1\*3 + 1\*5.

Из каждого числа k логично сначала вычитать максимально возможное слагаемое с участием 5, а потом уже проверять можно ли остаток добрать с помощью трёх шариков мороженого. Но нужно учесть, что, например, число 22 нельзя разложить как 4\*5 + n\*3 и нельзя разложить как 3\*5 + n\*3, но можно разложить как 2\*5 + 4\*3, поэтому нужно проверять все комбинации.

```python
k = int(input())
ans = False

for i in range(k // 5 + 1):
    res = k - 5 * i
    if res >= 0 and res % 3 == 0:
        ans = True
        break

print("Yes" if ans else "No")
```

На этом моменте пришло осознание, что любые числа, больше чем 7, можно разложить порциями по 3 и по 5, проверим на первых 100:

```python
ans = False

for k in range(1, 100):
    ans = False
    for i in range(k // 5 + 1):
        res = k - 5 * i
        if res >= 0 and res % 3 == 0:
            ans = True
            break
    
    print(k, "Yes" if ans else "No")
```
