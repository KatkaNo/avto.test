Не совсем поняла условие задачи, но додумаю так:
число k, которое подаётся на вход - это стоимость одного товара. Нужно рассчитать стоимость товаров в разном кол-ве: от 2 до 10.

```python
k = float(input())
n = 2

while n <= 10:
    price = k * n
    print(n, " price: ", price)
    n += 1
```

Если нужно хранить результат, а не только вывести на экран:

```python
k = float(input())
n = 2
result = []

while n <= 10:
    price = k * n
    result.append(str(price))
    n += 1

print(", ".join(result))
```
