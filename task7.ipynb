Первая идея - просто отдельно выбрать уникальных клиентов типа А и клиентов типа В. 
Затем можно ввести флаг, который окончательно будет определять тип клиента (после проверки принадлежности клиента обоим типам одновременно), и уже на его основе продолжать работу над целью запроса. Это можно сделать с помощью  case и подзапроса по условию равенства по полю client.

```python
select *,
    case 
        when exists (select client from clienttype ct 
             where ct.client = clienttype.client and ct.type = 'A') then 'A'
        else type
    end as final_type
from clienttype;
```

Теперь осталось выбрать и посчитать уникальные записи, опираясь на поля client и final_type:

```python
select 
    case 
        when exists (select client from clienttype ct 
             where ct.client = clienttype.client and ct.type = 'A') then 'A'
        else type
    end as final_type,
    count(distinct client) as N_clients
from clienttype
group by final_type;
```
