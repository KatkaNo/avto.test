RIGHT JOIN берёт каждую строку из таблицы, которая стоит после ключевого слова "JOIN", и ищет по условию связки соотв. ей строку в таблице, которая стоит после слова "FROM". 
При отсутствии совпадений строки "правой" таблицы дополняются null и включаются в результат.

**Запросы для проверки:**

```python
create table A (
Id int primary key,
Name varchar(50)
);


-- тут я поняла, что не хочу руками проставлять id, поэтому:

alter table A modify column Id int AUTO_INCREMENT primary key;


-- теперь выяснилось, что в SQLite, в котором я сейчас работаю, не поддерживается modify column, поэтому:

drop table A;


create table A (
Id integer primary key AUTOINCREMENT,
Name varchar(50)
);


insert into A (Name)
values 
    ('Alex'),
    ('Bob'),
    ('Alex'),
    ('Bob'),
    ('Ola'),
    ('Vika'),
    ('Alex'),
    ('Timur'),
    ('Mikael'),
    ('Kate');

    
select * from A;


create table B (
Id integer primary key AUTOINCREMENT,
Name varchar(50),
Phone integer
);


insert into B (Name, Phone)
values 
    ('Alex', 123),
    ('Bob', 234),
    ('Anton', 456),
    ('Vitya', 789),
    ('Max', 1012);


select * from B;

-- в итоге проверка задания:

SELECT * FROM A 
RIGHT JOIN B ON A.Name = B.Name;

-- а здесь выяснилось, что в SQLite нет еще и RIGHT JOIN (есть только LEFT OUTER JOIN), поэтому меняю таблицы местами.

SELECT * FROM B
LEFT OUTER JOIN A ON A.Name = B.Name;
```
